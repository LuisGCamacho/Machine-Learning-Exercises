## Simple Linear Regression: Wage Prediction.

### Statement and context of the exercise:

A technology company wants to analyze the relationship between an employee’s years of work experience and their annual salary. For this purpose, a dataset was collected containing the following variables:

YearsExperience: number of years of professional experience.
Salary: annual salary earned by the employee.

The objective is to study how salary changes according to the employee’s experience level and build a mathematical model capable of predicting salaries based on years of experience.

Using the provided dataset, the tasks are:

Visualize the relationship between both variables using a graph.
Build a predictive model using the data.
Obtain the equation of the line that best fits the data.
Predict the salary of employees based on their years of experience.
Analyze how accurately the model represents the relationship between experience and salary.

### 1. Import required libraries.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as ptl
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

### 2. Reading CSV data.

```python
dataset = pd.read_csv("datasets/Salary_Data.csv")
```

### 3. Dataset Exploration.

```python
dataset
```

```python
x = dataset.iloc[:, :-1]
y = dataset.iloc[:, 1]

print("Data x: \n", x)
print("Data y: \n", y)
```

### 3. Visualize dataset

```python
# Representación gráfica del conjunto de datos
plt.plot(x, y, "b.")
plt.xlabel("Years of professional experience")
plt.ylabel("Salary")
plt.show()
```

### 5. Divide the dataset into a training dataset and a test dataset.

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 1/6,
    random_state = 0,
    shuffle = True
)
```

### 6. Build Simple Linear Regression model.

```python
# Provide the model with the training data
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
# parameter theta 0
regression.intercept_
```

```python
# parameter theta 1
regression.coef_
```

```python
# Prediction for test dataset
y_predict = regression.predict(x_test)
```

```python
# Prediction for the minimum and maximum value of the training dataset
x_min_max = np.array([x.min(), x.max()])
y_train_pred = regression.predict(x_min_max)
```
