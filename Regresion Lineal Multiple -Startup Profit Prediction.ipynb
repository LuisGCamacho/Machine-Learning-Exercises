# Problem Statement

In today's competitive business environment, startups must decide how to allocate their financial resources among key areas such as Research and Development (R&D), Administration, and Marketing. Making the right investment decisions is essential for improving profitability and ensuring long-term business growth.

However, many companies invest significant amounts of money without a clear understanding of how these investments influence their future profits. As a result, inefficient budget allocation can reduce competitiveness and limit financial performance.

To address this issue, a predictive model can be developed to estimate a startup's profit based on its investment in different business areas and the state where it operates. Such a model provides valuable support for strategic planning and financial decision-making.

This project uses Multiple Linear Regression to predict the profit of startups because the target variable depends on several independent variables simultaneously.

### 0- Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### 1- DataSet Import

```python
dataset = pd.read_csv("datasets/50_Startups.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

### 4- Encoding Categorical Data


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(sparse_output=False), [3])
    ],
    remainder='passthrough'
)


X = ct.fit_transform(X)

X
```

### 4- Avoiding Multicollinearity by Dropping One Dummy Variable

```python
X = X[:, 1:]
X
```

### 6- Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    test_size = 0.20,
    random_state = 0,
    shuffle = True
)
```

### 7- Building the Optimal Multiple Linear Regression Model Using Backward Elimination

```python
import statsmodels.api as sm
```

```python
import numpy as np
import statsmodels.api as sm

# Add a column of ones (intercept)
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)


def backward_elimination(X, y, significance_level=0.05):

    # Convert to NumPy arrays
    X = np.asarray(X, dtype=float)
    y = np.asarray(y, dtype=float)

    # Store the current column indices
    columns = list(range(X.shape[1]))

    while True:

        # Build the regression model
        model = sm.OLS(y, X).fit()

        # Get the highest p-value
        max_p_value = max(model.pvalues)

        # If the highest p-value exceeds the significance level
        if max_p_value > significance_level:

            # Index of the least significant variable
            remove_index = model.pvalues.argmax()

            # Remove the corresponding column
            X = np.delete(X, remove_index, axis=1)

            # Remove the stored column index
            columns.pop(remove_index)
        else:
            break

    return X, model, columns


# Train the final model
X_train_final, model, columns = backward_elimination(X_train, y_train)

# Keep the same columns in the test set
X_test_final = X_test[:, columns]

# Make predictions
y_pred = model.predict(X_test_final)
```
